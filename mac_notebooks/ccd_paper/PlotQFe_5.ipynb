```python
def FOM(params):
    fom = 0.0
    [A, C1, C2, Beta, N0] = params
    B = -1.0/.014993
    for i,ne in enumerate(Ne):
        qfe = QFe[i]
        qfe_model = (ne / A + C1) * (1.0 + tanh(Beta*(ne-N0))/2.0) + (log(ne) / B + C2) * (1.0 - tanh(Beta*(ne-N0))/2.0)
        #print qfe, qfe_model
        if ne < 0.01:
            continue
        else:
            fom += (qfe - qfe_model)**2
    return fom
```
