```python
from lsst.summit.utils.efdUtils import makeEfdClient, getEfdData
import numpy as np
import matplotlib.pyplot as plt
from lsst.daf.butler import Butler
from astropy.time import Time, TimeDelta
```

```python
client = makeEfdClient()
```

```python
begin = Time("2025-04-19T02:00:00", scale='utc')
end = Time("2025-04-19T02:14:00", scale='utc')
expRecord = butlerUtils.getExpRecordFromDataId(butler, dataId)
zData = getEfdData(client=client, topic="lsst.sal.MTAOS.logevent_wavefrontError", 
                   begin=begin, end=end)
len(zData)
```

```python
detectors = [191, 195, 199, 203]
names = ['LL', 'LR', 'UL', 'UR']
numZernikes = 15
index = -1
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
for n, detector in enumerate(detectors):
    thisZData = zData[zData['sensorId'] == detector]
    xaxis = []
    yaxis = []
    for i in range(numZernikes):
        xaxis.append(int(thisZData[f'nollZernikeIndices{i}'].values[index]))
        yaxis.append(thisZData[f'nollZernikeValues{i}'].values[index])
    ax.plot(xaxis, yaxis, label=names[n])
ax.legend()
ax.set_xticks(xaxis)
timestamp = Time(zData['private_sndStamp'].values[index], scale='utc', format='unix_tai').isot.split('.')[0]
ax.set_title(
    f"Zernikes {timestamp}")
ax.set_xlabel("Noll Zernike Index")
ax.set_ylabel("Zernike value")
ax.set_ylim(-2.0, 2.0)
plt.savefig(f"/home/cslage/DATA/Zernike_{timestamp}.png")
```
