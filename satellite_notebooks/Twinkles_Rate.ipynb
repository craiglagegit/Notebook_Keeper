```python
import numpy as np
```

```python
r = 0.64 # m Radius of the camera body
L = 0.2 # m  This is the length of the camera body that can impact the detectors.
area = 2.0 * np.pi * r * L # m^2
print(f" Area = {area:.2f} m^2")
volume = area * 5.0E-6 # m^3  Only counting the top 5 microns
print(f" Volume = {volume:.3g} m^3")
mass = volume * 1.0E6 * 2.3  # grams  Al is 2.3 g/cm^3
print(f" Mass = {mass:.1f} grams of aluminum.")
N_al = mass / 26 * 6.02E23 # Aluminum is 26 g/mole
print(f" N_al = {N_al:.2g} aluminum atoms")
N_U = N_al * 1E-7
print(f"At 100ppb, N_U = {N_U:.2g} Uranium (or Thorium) atoms")
half_life = 4.5E9 * 86400 * 365.25 # sec half-life of Uranium
rate = np.log(2.0) / half_life # decays/atom/sec
decays_per_second = rate * N_U
print(f"So there are approximately {decays_per_second:.2f} decays_per_second")
decays_impacting_detector_per_image = decays_per_second * 0.25 * 30.0
print(f"Assume about 25% of these impact the detector, \nthen there are about {decays_impacting_detector_per_image:.1f} impacting the detector per image. \nThis is the right order of magnitude")
```
