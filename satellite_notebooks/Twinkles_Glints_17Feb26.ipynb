```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mSun = -26.7
mGlint = 8.5
Flux_ratio = 10**(-0.4 * (mGlint - mSun))
print(f"{Flux_ratio:.2g}")
```

```python
Rsun = 7.0E8 # meters
Dsun = 1.5E11 # meters
Dgeo = 3.6E7 # meters
```

```python
# pi Rgeo^2 / Dgeo^2 = Flux_ratio * pi Rsun^2 / Dsun^2

Rgeo = np.sqrt(Flux_ratio * Rsun**2 * Dgeo**2 / Dsun**2)
Rgeo * 100.0 # cm
```

```python
Rmoon = 1.7E6 # meters
Dmoon = 3.9E8 # meters
mMoon = -12.7
Moon_Flux_ratio = 10**(-0.4 * (mMoon - mSun))
Moon_Flux_ratio
```

```python
# Guillem's methodology
Moon_Albedo = 0.14

Amoon = Moon_Flux_ratio * np.pi * Dmoon**2 / Moon_Albedo
Rmoon_calc = np.sqrt(Moon_Flux_ratio * Dmoon**2 / Moon_Albedo)
print(f"Calculated moon radius = {Rmoon_calc/1000:.1f} km")
print(f"Actual moon radius = {Rmoon/1000:.1f} km")
```
