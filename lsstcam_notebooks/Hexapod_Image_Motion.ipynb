```python
import batoid
import numpy as np
```

```python
telescope = batoid.Optic.fromYaml("LSST_r.yaml")
```

```python
rays = batoid.RayVector.asPolar(
    optic=telescope,
    theta_x = 0.0,
    theta_y = 0.0,
    wavelength = 622e-9,
    nrad=10, naz=62
)
trays = telescope.trace(rays.copy())
trays = trays[~trays.vignetted]
print(f"{trays.x.mean():.9f}, {trays.y.mean():.9f}")
```

```python
perturbed = telescope.withGloballyShiftedOptic("LSSTCamera", [1E-6, 0.0, 0.0])
prays = perturbed.trace(rays.copy())
prays = prays[~prays.vignetted]
print(f"{prays.x.mean():.9f}, {prays.y.mean():.9f}")
```

```python
perturbed = telescope.withGloballyShiftedOptic("LSSTCamera", [0.0, 1E-6, 0.0])
prays = perturbed.trace(rays.copy())
prays = prays[~prays.vignetted]
print(f"{prays.x.mean():.9f}, {prays.y.mean():.9f}")
```

```python
perturbed = telescope.withGloballyShiftedOptic("M2", [1E-6, 0.0, 0.0])
prays = perturbed.trace(rays.copy())
prays = prays[~prays.vignetted]
print(f"{prays.x.mean():.9f}, {prays.y.mean():.9f}")
```

```python
perturbed = telescope.withGloballyShiftedOptic("M2", [0.0, 1E-6, 0.0])
prays = perturbed.trace(rays.copy())
prays = prays[~prays.vignetted]
print(f"{prays.x.mean():.9f}, {prays.y.mean():.9f}")
```
