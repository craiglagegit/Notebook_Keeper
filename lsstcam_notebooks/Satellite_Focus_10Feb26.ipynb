```python
import numpy as np
```

```python
def OutOfFocus(d1):
    # 1/f = 1/d1 + 1/d2  Simple lens equation
    # at infinity 1/d1=0 and d2 = f
    f = 8.3 * 1.2 #  Simonyi focal length in m
    d2 = 1.0 / (1.0 / f - 1.0 / (d1 * 1000.0))
    print(f"at a distance of {d1:.0f} km, we are {(f-d2) * 1E6:.0f} microns out of focus")
```

```python
OutOfFocus(400) # LEO
```

```python
OutOfFocus(40000) # GEO
```

```python
OutOfFocus(13000) # GEO
```
