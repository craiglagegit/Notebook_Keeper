```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from astropy.time import Time, TimeDelta
from lsst.daf.butler import Butler
from lsst.summit.utils.efdUtils import makeEfdClient, getEfdData
from lsst.summit.extras.slewTimingSimonyi import plotExposureTiming
from lsst.summit.utils.butlerUtils import getExpRecordFromDataId, makeDefaultButler
```

```python
client = makeEfdClient()
butler = makeDefaultButler('LSSTCam', embargo='False')
```

```python
%matplotlib inline
expId1 = 2026010900400
expId2 = 2026010900401
dataId1 = {'exposure':expId1, 'instrument':'LSSTCam'}
dataId2 = {'exposure':expId2, 'instrument':'LSSTCam'}
expRecord1 = getExpRecordFromDataId(butler, dataId1)
expRecord2 = getExpRecordFromDataId(butler, dataId2)
fig = plotExposureTiming(client, [expRecord1, expRecord2])
fig
fig.savefig(f"/home/c/cslage/u/MTMount/between_image_timing/Events_{expId1}_{expId2}.png", 
            bbox_inches='tight', pad_inches=1.2)
```
