```python
import numpy as np
import matplotlib.pyplot as plt
from lsst.daf.butler import Butler
from lsst.obs.lsst import LsstCam
from lsst.summit.utils.plotting import plot
```

```python
butler = Butler('/repo/embargo', collections=['LSSTCam/raw/all', 
                                            'LSSTCam/runs/nightlyValidation',
                                              'LSSTCam/raw/guider'])
instrument = 'LSSTCam'
camera = LsstCam.getCamera()
```

```python
expId  = 2026022400498
detName = 'R40_SG0'
stamp = 12
detector = camera[detName]
raw = butler.get("guider_raw", exposure=expId, detector=detector.getId(), instrument='LSSTCam')
mdata = raw.metadata
cols = mdata['ROICOLS']
rows = mdata['ROIROWS']
llX = mdata['ROICOL']
llY = mdata['ROIROW']
urX = llX + mdata['ROICOLS']
urY = llY + mdata['ROIROWS']
#stamp_ampName = ampDict[mdata['ROISEG']]
#print(stamp_ampName)

arr = raw[stamp].stamp_im.image.array
print(np.min(arr), np.max(arr), np.median(arr))
print(cols, rows, llX, llY, urX, urY)
```

```python
%matplotlib inline
arr = arr - np.median(arr)
fig = plot(arr)
ax = fig.get_axes()[0]
ax.set_title(f"{expId}_{detName}_{stamp}")
fig.savefig(f"/home/c/cslage/u/Guider_Mode/images/{expId}_{detName}_{stamp}.png")
```
